```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# model set-up
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
# setting up input strings from file
with open('input-text.txt') as f:
    input = f.readlines()

#sequence = "eating a guava"

sequence = ''
for elem in input: 
   sequence += ' ' + elem

print(sequence)
```

```python
# encoding the inputs
inputs = tokenizer.encode(sequence, return_tensors='pt')
```

```python
# generating outputs w/ model
outputs = model.generate(inputs, max_length=550, do_sample=True)
```

```python
# decoding the numerical output into actual text
text = tokenizer.decode(outputs[0], skip_special_tokens=True, temperature=3) 
# temperature parameter can be used to control randomness, the higher it is, the more random the text will be
print(text)
```

```python
# write output text into a mis-trans.txt file
with open('mis-trans.txt','w') as f:
    f.writelines(text)
```
